Load the Iris dataset and perform exploratory data analysis using visualizations.

## Load the dataset

### Subtask:
Load the Iris dataset into a pandas DataFrame.


**Reasoning**:
The instructions require loading the Iris dataset and converting it into a pandas DataFrame. This can be achieved by importing the necessary function from scikit-learn, loading the dataset, creating a DataFrame using the data and feature names, and adding the target variable.



```python
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Create a pandas DataFrame
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Add the target variable as a new column
df['species'] = iris.target

# Display the first 5 rows
display(df.head())
```

## Explore the dataset

### Subtask:
Display basic information about the dataset, such as the first few rows, column names, and data types.


**Reasoning**:
Display basic information about the dataset as requested by the subtask.



```python
df.info()
display(df.describe())
```

## Create visualizations

### Subtask:
Generate various plots, such as scatter plots, histograms, and box plots, to visualize the relationships between different features in the dataset.


**Reasoning**:
Import the necessary libraries for plotting and generate scatter plots, histograms, and box plots as requested in the instructions.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a scatter plot of 'sepal length (cm)' vs. 'sepal width (cm)', colored by 'species'.
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='sepal length (cm)', y='sepal width (cm)', hue='species')
plt.title('Scatter Plot of Sepal Length vs. Sepal Width')
plt.show()

# Create a scatter plot of 'petal length (cm)' vs. 'petal width (cm)', colored by 'species'.
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='petal length (cm)', y='petal width (cm)', hue='species')
plt.title('Scatter Plot of Petal Length vs. Petal Width')
plt.show()

# Generate histograms for each of the four features
df.hist(figsize=(10, 8))
plt.suptitle('Histograms of Iris Features', y=1.02)
plt.tight_layout()
plt.show()

# Create box plots for each of the four features, grouped by 'species'
features = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
plt.figure(figsize=(12, 10))
for i, feature in enumerate(features):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(data=df, x='species', y=feature)
    plt.title(f'Box Plot of {feature} by Species')
plt.tight_layout()
plt.show()
```
